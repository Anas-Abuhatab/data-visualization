```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

```

```python
taxis=sns.load_dataset("taxis")
taxis.head()
```

```python
sns.relplot(x="distance",y="total",hue="tolls",data=taxis)
```

### - we can clearly see above that there is direct relationship between distance and the total cost 

```python
f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.scatterplot(x="passengers", y="tip",
                hue="fare",
                sizes=(1,8),
                data=taxis)
```

### - we can clearly see above that the avarage best tip when taxi have one passenger 

```python
taxis_color= taxis.groupby("color" ,as_index=False).sum()
taxis_color
```

```python
f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.barplot(data=taxis_color, x = "color", y = "passengers" ,palette="YlOrBr_r",hue="passengers",)
```

### - we can clearly see above that the passengers prefare the yellow Taxi

# Conclusions:

* We can conclude that there is a linear relationship between distance and the total cost 
* Taxi well get more tips when have one passenger
