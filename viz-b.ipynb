```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("accidents_2012_to_2014.csv")
data.head()
```

```python
data_Year = data.groupby("Year" ,as_index=False).sum()
f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.barplot(data=data_Year, x = "Year", y = "Number_of_Vehicles" ,palette="YlOrBr_r",hue="Number_of_Vehicles",)
```

### - Number of Taxi accident in 2012,2013,2014 because taxi always in the road

```python
sns.relplot(data=data, x = "Road_Surface_Conditions", y = "Number_of_Casualties" ,palette="YlOrBr_r",hue="Number_of_Casualties",)
```

### - we can clearly see above that number of casualties is higher when road surface is Dry

```python
f, ax = plt.subplots(figsize=(10, 6.5))

sns.set_theme(style="whitegrid")
sns.lineplot(x="Light_Conditions", y="Number_of_Vehicles",data=data)

```

### - we can clearly see above that the lowest number of accident when Light Conditions is Day light street light present 

# Conclusions:

* Number of Taxi accident in 2012,2013,2014 because taxi always in the road
* Number of casualties is higher when road surface is Dry
* lowest number of accident when Light Conditions is Day light street light present
